```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/LLCP2023_clean.csv', keep_default_na=False, dtype='category')
```

```python
label_encoders = {}
for column in df.columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
```

```python
df = df.astype('category')
```

```python
df.info()
```

```python
import pycaret
pycaret.__version__
```

```python
from pycaret.classification import *
# initialize setup
s = setup(data = df, target = '_MENT14D', data_split_stratify=True, normalize = True, session_id=123)
```

```python
best = compare_models()
```

```python
print(best)
```

```python
tune_model(best)
```

```python
evaluate_model(best)
```

```python
plot_model(best, plot = 'auc')
```

```python
plot_model(best, plot = 'feature')
```

```python
plot_model(best, plot = 'confusion_matrix')
```

```python
predict_model(best)
```

```python
predictions = predict_model(best, data=df, raw_score=True)
predictions.head()
```
