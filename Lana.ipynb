```python
def label_Temp(row):
   if row['_DRNKWK2'] < 0.99:
      return 'No'
   else : return 'Yes'

df_lana.apply(label_Temp, axis=1)
```

```python
df_lana['_DRNKWK21'] = df_lana.apply(label_Temp, axis=1)
df_lana = df_lana.drop(['_DRNKWK2'], axis=1)
```

```python
sex_col = {
          1: 'Male',
          2: 'Female'
          }
df_lana['SEXVAR'] = df_lana['SEXVAR'].replace(sex_col)

hlth_col = {
          1: 'Excellent',
          2: 'Very good',
          3: 'Good',
          4: 'Fair',
          5: 'Poor'
          }
df_lana['GENHLTH'] = df_lana['GENHLTH'].replace(hlth_col)

phlth_col = {
          88: 'None'
          }
df_lana['POORHLTH'] = df_lana['POORHLTH'].replace(phlth_col)

exer_col = {
          1: 'Yes',
          2: 'No'
          }
df_lana['CVDSTRK3'] = df_lana['CVDSTRK3'].replace(exer_col)
df_lana['CHCCOPD3'] = df_lana['CHCCOPD3'].replace(exer_col)
df_lana['ADDEPEV3'] = df_lana['ADDEPEV3'].replace(exer_col)
df_lana['CHCKDNY2'] = df_lana['CHCKDNY2'].replace(exer_col)
df_lana['_MICHD'] = df_lana['_MICHD'].replace(exer_col)
df_lana['_DRDXAR2'] = df_lana['_DRDXAR2'].replace(exer_col)
df_lana['_TOTINDA'] = df_lana['_TOTINDA'].replace(exer_col)
df_lana['CHCSCNC1'] = df_lana['CHCSCNC1'].replace(exer_col)
df_lana['CHCOCNC1'] = df_lana['CHCOCNC1'].replace(exer_col)


diabete_col = {
          1: 'Yes',
          2: 'Yes',
          3: 'No',
          4: 'Yes'
          }
df_lana['DIABETE4'] = df_lana['DIABETE4'].replace(diabete_col)

bmi_col = {
          1: '77',
          2: 'Normal Weight',
          3: 'Overweight',
          4: 'Obese'
          }
df_lana['_BMI5CAT'] = df_lana['_BMI5CAT'].replace(bmi_col)

sad_col = {
          1: 'Always',
          2: 'Usually',
          3: 'Sometimes',
          4: 'Rarely',
          5: 'Never'
          }
df_lana['SDLONELY'] = df_lana['SDLONELY'].replace(sad_col)
df_lana['SDHSTRE1'] = df_lana['SDHSTRE1'].replace(sad_col)

race_col = {
          1: 'White',
          2: 'Other',
          3: 'Other',
          4: 'Other',
          5: 'Other',
          6: 'Other'
          }
df_lana['_IMPRACE'] = df_lana['_IMPRACE'].replace(race_col)

phy_col = {
          1: 'Zero days',
          2: '1-13 days',
          3: '	14+ days'
          }
df_lana['_PHYS14D'] = df_lana['_PHYS14D'].replace(phy_col)
df_lana['_MENT14D'] = df_lana['_MENT14D'].replace(phy_col)

hbp_col = {
          1: 'No',
          2: 'Yes'
          }
df_lana['_RFHYPE6'] = df_lana['_RFHYPE6'].replace(hbp_col)
df_lana['_RFCHOL3'] = df_lana['_RFCHOL3'].replace(hbp_col)

ast_col = {
          1: 'Yes',
          2: 'Yes',
          3: 'No'
          }
df_lana['_ASTHMS1'] = df_lana['_ASTHMS1'].replace(ast_col)

age_col = {
          1: 'Age 18 to 29',
          2: 'Age 18 to 29',
          3: 'Age 30 to 39',
          4: 'Age 30 to 39',
          5: 'Age 40 to 49',
          6: 'Age 40 to 49',
          7: 'Age 50 to 59',
          8: 'Age 50 to 59',
          9: 'Age 60 to 69',
          10: 'Age 60 to 69',
          11: 'Age 70+',
          12: 'Age 70+',
          13: 'Age 70+',
          14: '7'
          }
df_lana['_AGEG5YR'] = df_lana['_AGEG5YR'].replace(age_col)

chld_col = {
          1: 'No',
          2: 'Yes',
          3: 'Yes',
          4: 'Yes',
          5: 'Yes',
          6: 'Yes'
          }
df_lana['_CHLDCNT'] = df_lana['_CHLDCNT'].replace(chld_col)

edu_col = {
          1: 'No',
          2: 'High School',
          3: 'Attended College',
          4: 'College or Technical School'
          }
df_lana['_EDUCAG'] = df_lana['_EDUCAG'].replace(edu_col)

smoker_col = {
          1: 'Yes',
          2: 'Yes',
          3: 'Yes',
          4: 'No'
          }
df_lana['_SMOKER3'] = df_lana['_SMOKER3'].replace(smoker_col)

smoker_col = {
          1: 'Yes',
          2: 'Yes',
          3: 'Yes',
          4: 'No'
          }
df_lana['_SMOKER3'] = df_lana['_SMOKER3'].replace(smoker_col)

age_col = {
          1: 'Less than $15,000',
          2: '$15,000 to < $25,000',
          3: '$25,000 to < $35,000',
          4: '$35,000 to < $50,000',
          5: '$50,000 to < $100,000',
          6: '$100,000 to < $200,00',
          7: '$200,000+',
          9: '77'
          }
df_lana['_INCOMG1'] = df_lana['_INCOMG1'].replace(age_col)

df_lana = df_lana[~df_lana[['GENHLTH', 'CVDSTRK3', 'CHCSCNC1', 'CHCOCNC1', 'CHCCOPD3', 'ADDEPEV3', 'CHCKDNY2', 'DIABETE4', 'SDLONELY', 'SDHSTRE1', '_PHYS14D', '_MENT14D', '_RFHYPE6', '_RFCHOL3', '_MICHD', '_ASTHMS1', '_AGEG5YR', '_CHLDCNT', '_EDUCAG', '_SMOKER3', '_TOTINDA']].isin([7, 9, '7']).any(axis=1)]
df_lana = df_lana[~df_lana[['POORHLTH', '_BMI5CAT', '_INCOMG1']].isin([77, 99, '77']).any(axis=1)]
```
